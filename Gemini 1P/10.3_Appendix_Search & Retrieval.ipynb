# Appendix 10.3: Search & Retrieval

Did you know you can use Gemini to **search through Wikipedia for you**? Gemini can find and retrieve articles, at which point you can also use Gemini to summarize and synthesize them, write novel content from what it found, and much more. And not just Wikipedia! You can also search over your own docs, whether stored as plain text or embedded in a vector datastore.

See our [RAG cookbook examples](https://github.com/googles/google-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb) to learn how to supplement Gemini's knowledge and improve the accuracy and relevance of Gemini's responses with data retrieved from vector databases, Wikipedia, the internet, and more. There, you can also learn about how to use certain [embeddings](https://docs.google.com/gemini/docs/embeddings) and vector database tools.

If you are interested in learning about advanced RAG architectures using Gemini, check out our [Gemini 3 technical presentation slides on RAG architectures](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782).
